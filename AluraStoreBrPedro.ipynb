### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()

lista_lojas_df = [loja, loja2, loja3, loja4]
```

#1. Análise do faturamento


```python
def calcular_faturamento_total(lista_dfs):
    faturamento_por_loja = {}
    for i, df in enumerate(lista_dfs):
        nome_loja = f"Loja {i+1}"
        faturamento = df['Preço'].sum()
        faturamento_por_loja[nome_loja] = faturamento
    return faturamento_por_loja

faturamentos = calcular_faturamento_total(lista_lojas_df)
for loja, faturamento in faturamentos.items():
    print(f"{loja}: R$ {faturamento:,.2f}")
```

Com base apenas no faturamento, a Loja 1 é a que mais vende, enquanto a Loja 4 apresenta o menor resultado financeiro.

# 2. Vendas por Categoria


```python
def calcular_vendas_por_categoria(lista_dfs):
    vendas_categoria_por_loja = {}
    for i, df in enumerate(lista_dfs):
        nome_loja = f"Loja {i+1}"
        vendas_categoria = df['Categoria do Produto'].value_counts()
        vendas_categoria_por_loja[nome_loja] = vendas_categoria
    return vendas_categoria_por_loja

vendas_por_categoria = calcular_vendas_por_categoria(lista_lojas_df)
for loja, vendas_categoria in vendas_por_categoria.items():
    print(f"\n{loja}:")
    print(vendas_categoria)
```

Um padrão claro emerge: as categorias de móveis e eletrônicos são consistentemente as mais populares em todas as quatro lojas, seguidas de perto por brinquedos. Isso indica que o perfil de demanda dos clientes é bastante semelhante em todas as unidades.

A Loja 4, apesar do menor faturamento, não apresenta uma distribuição de categorias muito diferente das outras. Isso pode sugerir que o problema não está no tipo de produto que ela vende, mas talvez em outros fatores.

# 3. Média de Avaliação das Lojas

```python
def calcular_media_avaliacoes(lista_dfs):
    media_avaliacoes_por_loja = {}
    for i, df in enumerate(lista_dfs):
        nome_loja = f"Loja {i+1}"
        media = df['Avaliação da compra'].mean()
        media_avaliacoes_por_loja[nome_loja] = media
    return media_avaliacoes_por_loja

media_avaliacoes = calcular_media_avaliacoes(lista_lojas_df)
for loja, media in media_avaliacoes.items():
    print(f"{loja}: Média de {media:.2f} estrelas")
```

As avaliações são bastante consistentes e positivas em todas as lojas, com todas elas orbitando em torno de 4 estrelas. Isso indica um bom nível de satisfação geral.

No entanto, podemos notar que a Loja 1, apesar de ter o maior faturamento, possui a menor média de avaliação entre as quatro. A diferença é pequena, mas sugere que a experiência do cliente pode não ser tão boa quanto nas outras unidades. A Loja 3 se destaca com a maior satisfação do cliente.

Este é um dado importante, pois uma loja com alto faturamento mas com clientes menos satisfeitos pode ter problemas de sustentabilidade a longo prazo.

# 4. Produtos Mais e Menos Vendidos

```python
def analisar_produtos_vendidos(lista_dfs):
    produtos_vendidos_por_loja = {}
    for i, df in enumerate(lista_dfs):
        nome_loja = f"Loja {i+1}"
        contagem_produtos = df['Produto'].value_counts()
        produtos_vendidos_por_loja[nome_loja] = contagem_produtos
    return produtos_vendidos_por_loja

produtos_vendidos = analisar_produtos_vendidos(lista_lojas_df)

for loja, contagem in produtos_vendidos.items():
    print(f"\n{loja}:")
    print("\n5 Produtos MAIS Vendidos:")
    print(contagem.head(5))
    print("\n5 Produtos MENOS Vendidos:")
    print(contagem.tail(5))
    print("\n")
```

Vemos grandes inconsistências no desempenho de produtos específicos entre as lojas:

    Micro-ondas: É um dos mais vendidos nas Lojas 1 e 2, mas um dos menos vendidos na Loja 3.

    Guarda roupas: É um dos mais vendidos na Loja 1, mas um dos menos vendidos na Loja 4.

    Mesa de jantar: Vende muito bem na Loja 3, mas mal na Loja 2.

Isso sugere que o problema de uma loja pode não ser a falta de demanda geral, mas sim uma má gestão de estoque ou estratégias de marketing local ineficazes para certos produtos. A Loja 4, por exemplo, vende bem alguns móveis (Cama box, Cama king), mas falha em outros (Guarda roupas).

# 5. Frete Médio por Loja


```python
def calcular_frete_medio(lista_dfs):
    frete_medio_por_loja = {}
    for i, df in enumerate(lista_dfs):
        nome_loja = f"Loja {i+1}"
        media_frete = df['Frete'].mean()
        frete_medio_por_loja[nome_loja] = media_frete
    return frete_medio_por_loja

fretes_medios = calcular_frete_medio(lista_lojas_df)
for loja, frete in fretes_medios.items():
    print(f"{loja}: Custo médio de R$ {frete:,.2f} por envio")
```

A Loja 1, que possui o maior faturamento, também arca com o maior custo médio de frete. No outro extremo, a Loja 4, de menor faturamento, é a que possui o frete mais baixo, em média.

Isso pode indicar diferentes perfis de produtos sendo vendidos (produtos maiores/mais pesados na Loja 1) ou diferentes localizações geográficas dos clientes de cada loja.


# Visualização (Gráficos)

```python
# Gráfico 1: Faturamento Total por Loja (Gráfico de Barras)
faturamento_ordenado = sorted(faturamentos.items(), key=lambda item: item[1], reverse=True)
lojas_faturamento = [item[0] for item in faturamento_ordenado]
valores_faturamento = [item[1] for item in faturamento_ordenado]

plt.figure(figsize=(10, 6))
bars = plt.bar(lojas_faturamento, valores_faturamento, color=['#4CAF50', '#8BC34A', '#CDDC39', '#FFC107'])
plt.title('Faturamento Total por Loja', fontsize=16)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Faturamento (em R$)', fontsize=12)
plt.yticks(np.arange(0, max(valores_faturamento) * 1.1, 200000))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'R$ {int(x/1000)}k'))
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 15000, f'R$ {yval/1000:.1f}k', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('grafico_faturamento.png')
print("Gráfico de faturamento salvo como 'grafico_faturamento.png'")
```

```python
# Gráfico 2: Avaliação Média por Loja (Gráfico de Barras Horizontais)
lojas_avaliacoes = list(media_avaliacoes.keys())
valores_avaliacoes = list(media_avaliacoes.values())

plt.figure(figsize=(10, 6))
bars = plt.barh(lojas_avaliacoes, valores_avaliacoes, color=['#2196F3', '#03A9F4', '#00BCD4', '#009688'])
plt.title('Avaliação Média dos Clientes por Loja', fontsize=16)
plt.xlabel('Média de Avaliação (de 1 a 5)', fontsize=12)
plt.ylabel('Lojas', fontsize=12)
plt.xlim(3.9, 4.1)
plt.grid(axis='x', linestyle='--', alpha=0.7)

for index, value in enumerate(valores_avaliacoes):
    plt.text(value + 0.005, index, f'{value:.2f}', va='center')

plt.tight_layout()
plt.savefig('grafico_avaliacoes.png')
print("Gráfico de avaliações salvo como 'grafico_avaliacoes.png'")
```

```python
# Gráfico 3: Distribuição de Vendas por Categoria (Gráficos de Pizza)
vendas_loja_1 = vendas_por_categoria['Loja 1']
vendas_loja_4 = vendas_por_categoria['Loja 4']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
fig.suptitle('Distribuição de Vendas por Categoria', fontsize=20)

# Pizza para Loja 1 (Maior Faturamento)
ax1.pie(vendas_loja_1, labels=vendas_loja_1.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
ax1.set_title('Loja 1 (Maior Faturamento)', fontsize=14)
centre_circle = plt.Circle((0,0),0.70,fc='white')
ax1.add_artist(centre_circle)


# Pizza para Loja 4 (Menor Faturamento)
ax2.pie(vendas_loja_4, labels=vendas_loja_4.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
ax2.set_title('Loja 4 (Menor Faturamento)', fontsize=14)
centre_circle = plt.Circle((0,0),0.70,fc='white')
ax2.add_artist(centre_circle)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('grafico_categorias.png')
print("Gráfico de categorias salvo como 'grafico_categorias.png'")
```

# Relatório de Análise de Desempenho

1. Introdução

Conforme solicitado, foi realizada uma análise aprofundada dos dados de vendas das quatro lojas da rede Alura Store. O objetivo deste relatório é identificar a unidade com o menor desempenho geral para auxiliá-lo na decisão estratégica de qual loja vender para investir em um novo empreendimento. A análise foi baseada em métricas chave, incluindo faturamento total, vendas por categoria, satisfação do cliente, desempenho de produtos específicos e custo médio de frete.

2. Análise de Desempenho Comparativo

Para determinar a performance de cada loja, foram avaliados os seguintes indicadores:

**Faturamento Total**

O faturamento é o principal indicador de saúde financeira. A análise revelou uma clara disparidade no desempenho de vendas entre as lojas.

    Loja 1: R$ 1.534.509,12

    Loja 2: R$ 1.488.459,06

    Loja 3: R$ 1.464.025,03

    Loja 4: R$ 1.384.497,58

Como o gráfico ilustra, a Loja 4 apresenta o menor faturamento, ficando significativamente abaixo das demais e se destacando como a de pior desempenho financeiro.

**Satisfação do Cliente (Avaliação Média)**

A percepção do cliente é vital para a sustentabilidade do negócio. Todas as lojas mantêm um bom padrão de qualidade, com avaliações médias próximas de 4 de 5 estrelas.

Ainda que a diferença seja pequena, a Loja 1, apesar de ser a que mais fatura, possui a menor média de satisfação (3.98). A Loja 4, por sua vez, possui uma avaliação de 4.00, indicando que a causa de seu baixo faturamento provavelmente não está relacionada a uma má reputação com os clientes.

**Análise de Vendas e Frete**

Ao analisarmos a composição das vendas, notamos que todas as lojas têm um perfil de consumo semelhante, com móveis e eletrônicos sendo as categorias mais populares. A análise da Loja 1 (maior faturamento) e da Loja 4 (menor faturamento) mostra que a estratégia de categorias não é o fator decisivo para a diferença de desempenho.

Um ponto relevante é o custo do frete médio:

    A Loja 1 possui o frete mais caro (R$ 34,69).

    A Loja 4 possui o frete mais barato (R$ 31,28).

Isso indica que a Loja 4 possui uma vantagem de custo que, no entanto, não está sendo convertida em vendas, reforçando a ideia de uma ineficiência operacional ou de mercado.

3. Conclusão e Recomendação

Com base na análise completa dos dados, a recomendação é que o Senhor João deve vender a Loja 4.

Esta decisão é fundamentada nos seguintes pontos críticos:

    Menor Desempenho Financeiro: A Loja 4 gera o menor faturamento da rede, sendo o argumento mais forte para sua descontinuação. O objetivo de vender uma loja é levantar capital, e faz sentido estratégico se desfazer do ativo de menor performance.

    Ineficiência Operacional: Apesar de possuir vantagens competitivas, como o menor custo de frete da rede e uma avaliação de clientes positiva, a loja não consegue transformar esses pontos fortes em resultados de vendas. Isso sugere problemas mais profundos, seja na gestão de estoque, na estratégia de marketing local ou na adaptação ao mercado em que está inserida.

    Desempenho Estável das Outras Lojas: As Lojas 1, 2 e 3 apresentam um faturamento consideravelmente maior e mais coeso entre si, formando um núcleo sólido para a rede. Embora a Loja 1 tenha uma avaliação ligeiramente inferior, seu alto faturamento a torna um ativo valioso que deve ser mantido e melhorado, não vendido.

A venda da Loja 4 permitirá ao Senhor João obter o capital necessário para seu novo empreendimento, ao mesmo tempo em que otimiza a rede Alura Store, concentrando esforços nas unidades de maior retorno.
